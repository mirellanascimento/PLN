```python
df = pd.read_csv("https://raw.githubusercontent.com/viniciusrpb/cic0269_natural_language_processing/main/datasets/corpora/tweets_polarity_classification/twitter-2013train-A.txt", sep='\t', encoding = 'utf-8', names=['id','polarity','text'])
df
```

```python
text_df = df['text']
text_df
```

```python
# Tornar texto minúsculo
df['preprocessed_text'] = text_df.str.lower()
df
```

```python
# Fução para remover padrões
def remove_pattern(input_text, pattern):
  r = re.findall(pattern, input_text)
  for word in r:
    input_text = re.sub(word, "", input_text)

  return input_text  
```

```python
# Remover os rt
df['preprocessed_text'] = np.vectorize(remove_pattern)(df['preprocessed_text'], 'rt[\s]+') 
df
```

```python
# Remover @usuarios
df['preprocessed_text'] = np.vectorize(remove_pattern)(df['preprocessed_text'], '@[A-Za-z0-9]+') 
df
```

```python
# Remover links sites
df['preprocessed_text'] = np.vectorize(remove_pattern)(df['preprocessed_text'], 'http?:\/\/') 
df
```

```python
# Remover caracteres especiais, pontuação e numeros. Inclusive hashtag #
df['preprocessed_text'] = df['preprocessed_text'].str.replace('[^A-Za-z]', ' ') 
df
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
stop_words
```

```python
# Remover stop_words
df['preprocessed_text'] = df['preprocessed_text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
df
```

```python
# Tokenizing dataframe

tokens = df['preprocessed_text'].apply(lambda x: x.split())
tokens
```

```python
# Lemmatization
nltk.download('omw-1.4')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
# Lemmatization
lemmatizer = WordNetLemmatizer()
df['preprocessed_text'] = df['preprocessed_text'].apply( lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
df
```
